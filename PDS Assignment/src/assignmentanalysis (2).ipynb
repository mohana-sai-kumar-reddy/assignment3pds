```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python

# Replace 'diabetes.csv' with your actual dataset file
data = pd.read_csv('/content/raw_data/diabetes.csv')

```

```python

glucose_data = data['Glucose']
np.random.seed(42)

```

```python
# Take a random sample of 25
sample = glucose_data.sample(n=25)

sample_mean = sample.mean()
sample_highest = sample.max()
population_mean = glucose_data.mean()
population_highest = glucose_data.max()
```

```python
 #Create a comparison bar chart
labels = ['Population Mean', 'Population Highest', 'Sample Mean', 'Sample Highest']
values = [population_mean, population_highest, sample_mean, sample_highest]

plt.bar(labels, values, color=['blue', 'green', 'orange', 'red'])
plt.ylabel('Glucose Values')
plt.title('Comparison of Sample and Population Glucose Statistics')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
# adding the dataset
data = pd.read_csv('/content/raw_data/diabetes.csv')
```

```python
bmi_data = data['BMI']
```

```python
# Calculate the 98th percentile
population_98th_percentile = np.percentile(bmi_data, 98)
```

```python
# Setting seed value
np.random.seed(42)
```

```python

sample = np.random.choice(bmi_data, size=25, replace=False)
```

```python

sample_98th_percentile = np.percentile(sample, 98)

# Create a comparison chart
labels = ['Population 98th Percentile', 'Sample 98th Percentile']
values = [population_98th_percentile, sample_98th_percentile]

plt.bar(labels, values, color=['blue', 'orange'])
plt.ylabel('BMI Values')
plt.title('Comparison of 98th Percentile of BMI for Sample and Population')
plt.show()
```
