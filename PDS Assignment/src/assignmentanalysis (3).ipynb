```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
# adding dataset
data = pd.read_csv('/content/raw_data/diabetes.csv')
```

```python
# need to extract the bloodpressure column
population_blood_pressure = data['BloodPressure']
```

```python
# Calculate population
population_mean = population_blood_pressure.mean()
population_std = population_blood_pressure.std()
population_percentile = np.percentile(population_blood_pressure, 95)
```

```python

num_samples = 500
sample_size = 150

bootstrap_means = [np.mean(np.random.choice(population_blood_pressure, size=sample_size, replace=True)) for _ in range(num_samples)]
bootstrap_stds = [np.std(np.random.choice(population_blood_pressure, size=sample_size, replace=True)) for _ in range(num_samples)]
bootstrap_percentiles = [np.percentile(np.random.choice(population_blood_pressure, size=sample_size, replace=True), 95) for _ in range(num_samples)]

```

```python
# Creating charts
plt.figure(figsize=(12, 4))

```

```python
# Comparison of means
plt.subplot(1, 3, 1)
plt.hist(bootstrap_means, bins=30, alpha=0.5, label='Bootstrap Means')
plt.axvline(x=population_mean, color='red', linestyle='dashed', linewidth=2, label='Population Mean')
plt.xlabel('Mean Blood Pressure')
plt.ylabel('Frequency')
plt.legend()
plt.title('Comparison of Means')

# Comparison of standard deviations
plt.subplot(1, 3, 2)
plt.hist(bootstrap_stds, bins=30, alpha=0.5, label='Bootstrap STDs')
plt.axvline(x=population_std, color='red', linestyle='dashed', linewidth=2, label='Population STD')
plt.xlabel('Standard Deviation of Blood Pressure')
plt.ylabel('Frequency')
plt.legend()
plt.title('Comparison of Standard Deviations')

# Comparison of percentiles
plt.subplot(1, 3, 3)
plt.hist(bootstrap_percentiles, bins=30, alpha=0.5, label='Bootstrap Percentiles')
plt.axvline(x=population_percentile, color='red', linestyle='dashed', linewidth=2, label='Population Percentile')
plt.xlabel('95th Percentile of Blood Pressure')
plt.ylabel('Frequency')
plt.legend()
plt.title('Comparison of Percentiles')

plt.tight_layout()
plt.show()
```
