```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python

# Replace 'diabetes.csv' with your actual dataset file
data = pd.read_csv('/content/raw_data/diabetes.csv')

```

```python

glucose_data = data['Glucose']
np.random.seed(42)

```

```python
# Take a random sample of 25
sample = glucose_data.sample(n=25)

sample_mean = sample.mean()
sample_highest = sample.max()
population_mean = glucose_data.mean()
population_highest = glucose_data.max()
```

```python
 #Create a comparison bar chart
labels = ['Population Mean', 'Population Highest', 'Sample Mean', 'Sample Highest']
values = [population_mean, population_highest, sample_mean, sample_highest]

plt.bar(labels, values, color=['blue', 'green', 'orange', 'red'])
plt.ylabel('Glucose Values')
plt.title('Comparison of Sample and Population Glucose Statistics')
plt.show()
```
